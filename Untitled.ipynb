```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import (
    HistGradientBoostingRegressor, 
    RandomForestRegressor, 
    ExtraTreesRegressor, 
    GradientBoostingRegressor,
    VotingRegressor
)
from sklearn.linear_model import Ridge, Lasso
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.metrics import mean_squared_error

# Configuration
pd.set_option('display.max_columns', None)
np.random.seed(42)
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries Loaded")
```

```python
#DATA LOADING
train = pd.read_csv("train.csv", parse_dates=['Date'], low_memory=False)
test = pd.read_csv("test.csv", parse_dates=['Date'], low_memory=False)
store = pd.read_csv("store.csv")

#Load External Data
doctor_visits = pd.read_csv("doctor_visits.csv", parse_dates=['Date'])
weather = pd.read_csv("weather.csv", parse_dates=['Date'])
holpol = pd.read_csv("holidays_policy.csv", parse_dates=['Date'])
adspend = pd.read_csv("ad_spend.csv", parse_dates=['Date'])

print(f"Data Loaded.")
print(f"   - Train Rows: {train.shape[0]:,}")
print(f"   - Test Rows:  {test.shape[0]:,}")
```

```python
#FEATURE ENGINEERING
#A.Store Cleaning
store['CompetitionDistance'] = store['CompetitionDistance'].fillna(store['CompetitionDistance'].median())
store[['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2SinceWeek', 'Promo2SinceYear']] = \
    store[['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2SinceWeek', 'Promo2SinceYear']].fillna(0)
store['PromoInterval'] = store['PromoInterval'].fillna("None")

#B.Region Mapping
n_regions = 6
unique_stores = sorted(store['Store'].unique())
regions_list = [f"R{i}" for i in range(1, n_regions+1)]
store_region_map = {s: regions_list[i % n_regions] for i, s in enumerate(unique_stores)}
store['Region'] = store['Store'].map(store_region_map)

#C.Date Features
def extract_date_features(df):
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['IsWeekend'] = df['DayOfWeek'].isin([5, 6]).astype(int)
    return df

train = extract_date_features(train)
test = extract_date_features(test)

#D.Duration Features
def calculate_durations(df, store_df):
    if 'CompetitionOpenSinceYear' not in df.columns:
        df = df.merge(store_df, on='Store', how='left')
    df['CompetitionOpen'] = 12 * (df.Year - df.CompetitionOpenSinceYear) + (df.Month - df.CompetitionOpenSinceMonth)
    df['CompetitionOpen'] = df['CompetitionOpen'].apply(lambda x: x if x > 0 else 0)
    df['Promo2Open'] = 12 * (df.Year - df.Promo2SinceYear) + (df.Week - df.Promo2SinceWeek) / 4.0
    df['Promo2Open'] = df['Promo2Open'].apply(lambda x: x if x > 0 else 0)
    return df

train = calculate_durations(train, store)
test = calculate_durations(test, store)

#E.Merge External Data
def merge_external(df):
    df = df.merge(weather, on=['Date','Store'], how='left')
    df = df.merge(adspend, on=['Date','Store'], how='left')
    df = df.merge(holpol, on=['Date'], how='left')
    df = df.merge(doctor_visits, on=['Date','Region'], how='left')
    return df

train = merge_external(train)
test = merge_external(test)

#F.Fill Missing Values
for f in ['DoctorVisits', 'MaxTemp', 'RainMM', 'AdSpend', 'HolidayFlag', 'PolicyFlag']:
    median_val = train[f].median()
    train[f] = train[f].fillna(median_val)
    test[f] = test[f].fillna(median_val)

print("Feature Engineering Complete")
```

```python
#EDA
#Sales Distribution
plt.figure(figsize=(10, 4))
sns.histplot(train['Sales'], bins=50, kde=True, color='teal')
plt.title("Distribution of Sales")
plt.show()

#Impact of External Factors (AdSpend vs Sales)
plt.figure(figsize=(10, 5))
# Sample data for clearer scatter plot
sample = train.sample(5000)
sns.scatterplot(x='AdSpend', y='Sales', hue='Promo', data=sample, palette='coolwarm', alpha=0.6)
plt.title("Ad Spend vs. Sales (Colored by Promo)")
plt.show()
```

```python
#PREPROCESSING & SPLITTING
#1.Filter Open Stores
train_clean = train[(train['Open'] == 1) & (train['Sales'] > 0)].copy().reset_index(drop=True)

train_clean['LogSales'] = np.log1p(train_clean['Sales'])

#2.Ordinal Encoding
cat_cols = ['StoreType', 'Assortment', 'StateHoliday', 'Region']
for c in cat_cols:
    train_clean[c] = train_clean[c].astype(str)
    test[c] = test[c].astype(str)

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
train_clean[cat_cols] = encoder.fit_transform(train_clean[cat_cols])
test[cat_cols] = encoder.transform(test[cat_cols])

#3.Define Features
features = [
    'Store', 'DayOfWeek', 'Promo', 'StateHoliday', 'SchoolHoliday',
    'StoreType', 'Assortment', 'CompetitionDistance', 'CompetitionOpen', 
    'Promo2', 'Promo2Open', 'Region',
    'DoctorVisits', 'MaxTemp', 'RainMM', 'AdSpend', 'HolidayFlag', 'PolicyFlag',
    'Year', 'Month', 'Day', 'Week', 'IsWeekend'
]

X = train_clean[features]
y = train_clean['LogSales']

#4.Scale Data
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=features)

#Scaling Test Data
test_scaled = pd.DataFrame(scaler.transform(test[features]), columns=features)

#5.Validation Split(Last 6 Weeks)
val_start_date = train_clean['Date'].max() - pd.Timedelta(weeks=6)
mask_train = train_clean['Date'] < val_start_date
mask_val = train_clean['Date'] >= val_start_date

X_train, y_train = X[mask_train], y[mask_train]
X_val, y_val = X[mask_val], y[mask_val]

print(f"Data Split & Scaled")
print(f"   - Training Set: {len(X_train)} rows")
print(f"   - Validation Set: {len(X_val)} rows")
```

```python
#MULTI-MODEL ENSEMBLE TRAINING
#Base Models
#set of algorithms to ensure the ensemble is robust.
# - HistGradientBoosting: Fast, state-of-the-art boosting (similar to LightGBM).
# - Random Forest & Extra Trees: Parallel bagging methods to reduce variance.
# - GradientBoosting: Classical sequential boosting.
# - Ridge & Lasso: Linear models to capture baseline trends and regularize.

estimators = [
    ('hist_gb', HistGradientBoostingRegressor(max_iter=200, random_state=42)),
    ('random_forest', RandomForestRegressor(n_estimators=50, max_depth=15, n_jobs=-1, random_state=42)),
    ('extra_trees', ExtraTreesRegressor(n_estimators=50, max_depth=15, n_jobs=-1, random_state=42)),
    ('gradient_boost', GradientBoostingRegressor(n_estimators=100, max_depth=5, random_state=42)),
    ('ridge', Ridge(alpha=1.0)),
    ('lasso', Lasso(alpha=0.01))
]

print(f"Training {len(estimators)} individual models + 1 Ensemble...")

#2.Creating Voting Regressor (The "Manager" Model)
#This averages the predictions of all the models above.
ensemble = VotingRegressor(estimators)

# 3.Fit the Ensemble
# This will automatically train all the sub-models inside it on the Training Data.
ensemble.fit(X_train, y_train)

print("Training Complete.")
```

```python
#EVALUATION, COMPARISON & SIMULATION
#1.Define Evaluation Metric (RMSPE)
def rmspe(y_true, y_pred):

    return np.sqrt(np.mean(((y_true - y_pred) / y_true) ** 2))

actual_sales = np.expm1(y_val)
results = []

#2.Evaluate Individual Models vs. The Ensemble
print("\nModel Leaderboard (Validation Set)")
for name, model in ensemble.named_estimators_.items():
    # Predict
    pred_log = model.predict(X_val)
    pred_sales = np.expm1(pred_log)
    
    # Score
    score = rmspe(actual_sales, pred_sales)
    results.append({'Model': name, 'RMSPE': score})
    print(f"   - {name}: {score:.4f}")

# Evaluate the Ensemble (Voting)
ensemble_pred_log = ensemble.predict(X_val)
ensemble_pred_sales = np.expm1(ensemble_pred_log)
ensemble_score = rmspe(actual_sales, ensemble_pred_sales)
results.append({'Model': '>> ENSEMBLE (Voting) <<', 'RMSPE': ensemble_score})

# Create DataFrame for Comparison
results_df = pd.DataFrame(results).sort_values('RMSPE')
print("\nFINAL RESULTS")
print(results_df)

#Graph A: Model Comparison Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x='RMSPE', y='Model', data=results_df, palette='viridis')
plt.title("Model Comparison: Lower RMSPE is Better")
plt.xlabel("RMSPE (Error)")
plt.show()

#Graph B: Forecast vs Actuals (Time Series)
#first 150 days to show how closely the ensemble follows the trend
plt.figure(figsize=(15, 6))
plt.plot(actual_sales.values[:150], label='Actual Sales', color='black', alpha=0.6, linewidth=2)
plt.plot(ensemble_pred_sales[:150], label='Ensemble Forecast', color='red', linestyle='--', linewidth=2)
plt.title("Ensemble Forecast vs Actuals (First 150 Validation Days)")
plt.legend()
plt.show()

#X-FACTOR ANALYSIS: Permutation Importance
print("\nCalculating Feature Importance (X-Factors)")
subset_idx = np.random.choice(len(X_val), size=2000, replace=False)
perm_result = permutation_importance(
    ensemble, X_val.iloc[subset_idx], y_val.iloc[subset_idx], 
    n_repeats=5, random_state=42
)
sorted_idx = perm_result.importances_mean.argsort()[::-1]

# Graph C: Feature Importance Bar Chart
plt.figure(figsize=(12, 6))
plt.bar(range(10), perm_result.importances_mean[sorted_idx[:10]], color='teal')
plt.xticks(range(10), [features[i] for i in sorted_idx[:10]], rotation=45, ha='right')
plt.title("The X-Factors: Top 10 Drivers of Sales")
plt.ylabel("Importance (Drop in Accuracy)")
plt.show()

#BUSINESS SIMULATION (What-If Analysis)
print("\nBusiness Simulation: Ad Spend ROI")
# Scenario: What if we increase AdSpend by 20%?
sim_data = X_val.iloc[subset_idx].copy()

#1.Baseline Revenue
base_pred = np.expm1(ensemble.predict(sim_data)).sum()

#2.Simulation: Increase AdSpend by 20%
sim_data['AdSpend'] = sim_data['AdSpend'] + 0.2 
new_pred = np.expm1(ensemble.predict(sim_data)).sum()

#3.Calculate Uplift
uplift = (new_pred - base_pred)
uplift_pct = (uplift / base_pred) * 100

print(f"Baseline Projected Revenue: ${base_pred:,.2f}")
print(f"Projected Revenue with Increased Ad Spend: ${new_pred:,.2f}")
print(f"Net Revenue Uplift: {uplift_pct:.2f}%")


#Predictions for the Test set
test_open = test.copy()
closed_mask = (test_open['Open'] == 0)

#Predict on Scaled Test Data
final_test_preds = np.expm1(ensemble.predict(test_scaled))
final_test_preds[closed_mask] = 0 # Force 0 for closed stores

submission = pd.DataFrame({'Id': test['Id'], 'Sales': final_test_preds})
submission.to_csv("submission_ensemble.csv", index=False)
```

```python
#PRESCRIPTIVE ANALYTICS & DEPLOYMENT
import joblib

print("Prescriptive Analysis")

#1.ELASTICITY ANALYSIS (Using the Linear Component)
print("\n--- Market Elasticity (Linear Sensitivity) ---")
ridge_model = ensemble.named_estimators_['ridge']
coefficients = ridge_model.coef_

# Map coefficients to feature names
elasticity_df = pd.DataFrame({'Feature': features, 'Coefficient': coefficients})
elasticity_df = elasticity_df.sort_values(by='Coefficient', ascending=False)

#Filter for key actionable drivers
key_drivers = ['Promo', 'AdSpend', 'DoctorVisits', 'CompetitionDistance']
print(elasticity_df[elasticity_df['Feature'].isin(key_drivers)])

#2.STRATEGIC OPTIMIZATION (Finding the "Sweet Spot")
print("\nBudget Optimization Simulation ---")
scale_factors = [0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5]
projected_revenues = []

# A subset for fast simulation
sim_subset = X_val.iloc[subset_idx].copy()
base_ad_spend = sim_subset['AdSpend'].copy()

for scale in scale_factors:
    sim_subset['AdSpend'] = base_ad_spend * scale 
    
    # Predict
    pred_log = ensemble.predict(sim_subset)
    total_rev = np.expm1(pred_log).sum()
    projected_revenues.append(total_rev)

#Find Optimal Scale
best_idx = np.argmax(projected_revenues)
best_scale = scale_factors[best_idx]
print(f"Optimal Ad Budget Scale: {best_scale*100}% of current spending.")
print(f"   (At this level, revenue is maximized)")

#Graph: Optimization Curve
plt.figure(figsize=(10, 5))
plt.plot(scale_factors, projected_revenues, marker='o', color='green', linewidth=2)
plt.axvline(best_scale, color='red', linestyle='--', label=f'Optimal ({best_scale}x)')
plt.title("Revenue Optimization: Finding the Ad Spend Sweet Spot")
plt.xlabel("Ad Spend Multiplier (1.0 = Current Budget)")
plt.ylabel("Projected Total Revenue")
plt.legend()
plt.grid(True)
plt.show()

#3.DEPLOYMENT
print("\nSaving Model for Production")
model_filename = 'PharmaIntelliX_Ensemble_v1.pkl'
joblib.dump(ensemble, model_filename)
print(f"Model saved as '{model_filename}'.")
```
